```python
"""
순차 탐색
리스트에 있는 첫 번째 자료부터 하나씩 비교하면서 같은 값이 나오면 그 위치를 결과로
돌려주고, 리스트 끝까지 찾아도 같은 값이 나오지 않으면 -1을 돌려주면 된다.
이 방법은 '리스트 안에 있는 원소를 하나씩 순차적으로 비교하면서 탐색한다' 라고 하여
'순차 탐색(sequential search)'라고 부른다.
"""
```

```python
"""
다음은 순차 탐색 알고리즘으로 리스트 [17, 92, 18, 33, 58, 5, 33, 42]에서 특정 값
(18, 33, 900)을 찾아서 해당 위치 번호를 돌려주는 프로그램이다.
"""
# 리스트에서 특정 숫자의 위치 찾기
# 입럭 : 리스트 a, 찾는 값 x
# 출력 : 찾으면 그 값의 위치, 찾지 못하면 -1

def search_list(a, x):
    n  = len(a)
    for i in range(0, n):
        if x == a[i]:
            return i
    return -1

v = [17, 92, 18, 33, 58, 7, 33, 42]
print(search_list(v, 18)) # 2(순서상 세 번째지만, 위치 번호는 2)
print(search_list(v, 33)) # 3(33은 리스트에 두 번 나오지만 처음 나온 위치만 출력)
print(search_list(v, 900)) # -1(900은 리스트에 없음)
```

```python
"""
찾는 값이 리스트의 맨 앞에 있다면 단 한 번만 비교해도 결과를 얻을 수 있다.
하지만 찾는 값이 리스트의 마지막에 있거나 아예 없다면 리스트의 끝까지 하나하나 비교해야 한다.
이 처럼 경우에 따라 계산 횟수가 다를 때는 최선, 평균, 최악의 경우로 나누어 각각 계산 복잡도를 생각해 보기도 한다.
이 중 최악의 경우를 분석하면 어떤 경우라도 그보다는 빨리 계산 할 수 있을 것이다.
따라서 보수적인 관점에서 이 알고리즘을 최악의 경우로 분석해보면 비교가 최대 n번 필요하고 계산 복잡도는 O(n)이다.
"""
```

```python
# 찾는 값이 나오는 모든 위치를 리스트로 돌려주는 탐색 알고리즘
def searching_list(a, x):
    n  = len(a)
    b = []
    for i in range(0, n):
        if x == a[i]:
             b.append(i) # 자료 i를 리스트의 맨 뒤에 추가한다.
    return b

v = [17, 92, 18, 33, 58, 7, 33, 42]
print(searching_list(v, 18)) # 2(순서상 세 번째지만, 위치 번호는 2)
print(searching_list(v, 33)) # 3(33은 리스트에 두 번 나오지만 처음 나온 위치만 출력)
```

```python
# 학생 번호에 해당하는 학생 이름 찾기

def search_stu(a, x, y):   
    n = len(a)
    z = '?'
    for i in range(0, n): 
        if y == a[i]: # 학생 번호가 찾는 학생번호와 같으면
            return x[i] # 해당하는 학생 이름을 결과로 반환
    
    return z 

stu_no = [39, 14, 67, 105]
stu_name = ["Justin", "John", "Mike", "Summer"]

print(search_stu(stu_no, stu_name, 39))
```

```python
"""
정렬(sort)란 자료를 크기 순서대로 맞춰 일렬로 나열하는 것이다.
리스트에 들어 있는 숫자를 크기 순으로 나열하는 정렬 알고리즘의 입출력은 다음과 같이 정리할 수 있다.

문제 : 리스트 안에 있는 자료를 순서대로 배열하기
입력 : 정렬할 리스트(예: [35, 9, 2, 85, 17])
출력 : 순서대로 정렬된 리스트(예 : [2, 9, 17, 35, 85])
"""
```

```python
# 입력 : 리스트 a
# 출력 : 정렬된 새 리스트
# 주어진 리스트에서 최소값의 위치를 돌려주는 함수

def find_min(a):
    n  = len(a)
    min = 0
    for i in range(1, n):
        if a[i] < a[min]:
            min = i
    return min

def sel_sort(a):
    result = [] # 새 리스트를 만들어 정렬된 값을 저장
    while a: # 주어진 리스트에 값이 남아 있는 동안 계속
        min = find_min(a) # 리스트에 남아 있는 값 중 최소값의 위치
        value = a.pop(min) # 찾은 최소값을 빼내어 value에 저장
        result.append(value) # value를 결과 리스트 끝에 추가
    return result

d = [2, 4, 5, 1, 3]
print(sel_sort(d))
```

```python
# 일반적인 선택 정렬 알고리즘
# 입력 : 리스트 a
# 출력 : 없음(입력으로 주어진 a가 정렬됨) 

def sel_sort(a):
    n = len(a)
    for i in range(0, n - 1): # 0부터 n - 2까지 반복
        # i번 위치부터 끝까지 자료 값 중 최소값의 위치를 찾음
        min = i
        for j in range(i + 1, n):
            if a[j] < a[min]:
                min = j
        a[i], a[min] = a[min], a[i] # 두 자료 값 서로 바꿈
        
d = [2, 4, 3, 5, 1]
sel_sort(d)
print(d)
```
