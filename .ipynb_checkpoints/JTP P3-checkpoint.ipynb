```python
a = "Life is too short, you need python"

if "wife" in a: print("wife")
elif "python" in a and "you" not in a: print("python")
elif "shirt" not in a: print("shirt") 
elif "need" in a: print("need") # 위에서 이미 if 조건이 돼서 스킵
else: print("none")
```

```python
number = 0
i = 1
while i < 1001:
    if i% 3 == 0:
        number += i
    i += 1
print(number)
```

```python
A = [70, 60, 55, 75, 95, 90, 80, 80, 85, 100]
total = 0

for score in A:
    total += score
    
ave = total / len(A)
print(ave)
```

```python
numbers = [1, 2, 3, 4, 5]
result = []
for n in numbers:
    if n % 2 == 1:
        result.append(n*2)
```

```python
result = [ n * 2 for n in numbers if n % 2 == 1]
print(result)
```
