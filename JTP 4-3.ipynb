```python
f = open("새(bird)파일.txt", 'w')
for i in range(1,11):
    data = "%d번째 줄. \n" % i
    f.write(data)
f.close()
```

```python
#readline 함수
f = open("새(bird)파일.txt", 'r')
line = f.readline()
print(line)
f.close()
```

```python
f = open("새(bird)파일.txt", 'r')
while True:
    line = f.readline()
    if not line: break
    print(line)
f.close()
```

```python
# readlines 함수
f = open("새(bird)파일.txt", 'r')
lines = f.readlines()
for line in lines:
     # line = line.strip()   < 줄 바꿈 제거
    print(line)
f.close()
```

```python
# read 함수
f = open("새(bird)파일.txt", 'r')
data = f.read() # 파일의 내용 전체를 문자열로 돌려줌.
print(data)
f.close()
```

```python
# 파일 내용 추가
f = open("새(bird)파일.txt", 'a')
for i in range(11, 20):
    data = "%d번째 줄. \n" % i
    f.write(data)
f.close()
```

```python
# with문
f = open("foo.txt", 'w')
f.write("niheeheee")
f.close()
```

```python
# with문을 사용하면 with블록을 벗어나는 순간 파일 객체가 자동으로 닫힘
with open("foo.txt", 'w') as f:
    f.write("niheeheee")
```
