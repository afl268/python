```python
# {Key1:Value1, Key2:Value2, Key3:Value3, ...}
# 딕셔너리 Key와 Value의 쌍 여러 개가 { }로 둘러싸여 있다. 
# 각각의 요소는 Key : Value 형태로 이루어져 있고 쉼표(,)로 구분되어 있다.

dic = {'name':'pey', 'phone':'0119993323', 'birth': '1118'}
# key = name, phone, birth        value = pey , 0119993323 , 1118

a = {1 : 'hi'}
a = { 'a' : [1,2,3]}
```

```python
# 딕셔너리 쌍 추가
a = {1: 'a'}
a[2] = 'b'
a
```

```python
a['name'] = 'pey'
a
```

```python
a[3] = [1, 2, 3]
a
```

```python
# 딕셔너리 요소 삭제
del a[1]
a
```

```python
# 딕셔너리에서 key 사용해 value 얻기
grade = {'pey' : 10, 'julliet' : 99}
grade['pey']
```

```python
grade['julliet']
# 'pey'라는 Key의 Value를 얻기 위해 grade['pey']를 사용한 것처럼 
# 어떤 Key의 Value를 얻기 위해서는 딕셔너리변수이름[Key]를 사용한다.
```

```python
# key 리스트 만들기(keys)
a = {'name':'pey', 'phone':'0119993323', 'birth': '1118'}
a.keys()
```

```python
for k in a.keys():
    print(k)
```

```python
list(a.keys())
```

```python
# value 리스트 만들기(values)
a.values()
```

```python
# key, value 쌍 얻기(items)
a.items()
```

```python
# key : value 쌍 모두 지우기(clear)
a.clear()
a
```

```python
# key로 value얻기(get)
a = {'name' : 'pey', 'phone' : '0119993323', 'birth' : '1118'}
a.get('name')
```

```python
a.get('phone')
```

```python
a = {'name' : 'pey', 'phone' : '0119993323', 'birth' : '1118'}
print(a.get('nokey'))
```

```python
print(a['nokey'])
```

```python
a.get('foo', 'bar') # key 값이 없을 경우 미리 정해 둔 디폴트 값 대신 가져오기
```

```python
# 해당 key가 딕셔너리 안에 있는지 조사(in)
a = {'name' : 'pey', 'phone' : '0119993323', 'birth' : '1118'}
'name' in a
```

```python
'email' in a
```
