```python
a = [(1, 2), (3, 4), (5, 6)]
for (first, last) in a:
    print(first + last)
```

```python
#"총 5명의 학생이 시험을 보았는데 시험 점수가 60점이 넘으면 합격이고 그렇지 않으면 불합격이다. 
# 합격인지 불합격인지 결과를 보여 주시오."

list = [90, 30, 40 ,100, 35]
number = 0
for human in list:
    number = number + 1
    if human >= 60:
        print("%d번 학생은 합격." % number)
    else:
        print("%d번 학생은 불합격." % number)
```

```python
# for문과 continue
list = [90, 30, 40 ,100, 35]

number = 0
for human in list:
    number = number + 1
    if human < 60:
        continue
    print("%d번 학생은 합격." % number)
```

```python
add = 0
for i in range(1, 11):
    add = add + i
    
print(add)
```

```python
list = [90, 30, 40 ,100, 35]

for number in range(len(list)):
    if list[number] < 60:
        continue
    print("%d번 학생 합격." % (number + 1))
```

```python
# 구구단
for i in range(2,10):
    for j in range(1, 10):
        print(i*j, end = " ")
    print('')
        
```

```python
# 리스트 내포 
a = [1, 2, 3, 4]
result = []
for num in a:
    result.append(num*3)
    

print(result)
```

```python
a = [1, 2, 3, 4]
result = [num * 3 for num in a]
print(result)
```

```python
# 만약 [1,2,3,4] 중에서 짝수에만 3을 곱하여 담고 싶다면 다음과 같이 리스트 내포 안에 "if 조건"을 사용
a = [1, 2, 3, 4]
result = [num * 3 for num in a if num % 2 == 0]
print(result)
```

```python
# 구구단
result = [x*y for x in range(2,10)
            for y in range(1, 10)]
print(result)
```
