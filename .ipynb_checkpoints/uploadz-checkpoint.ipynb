```python
import pandas as pd
```

```python
series = pd.Series([17000, 20000, 2000, 4000],
               index=["피자", "치킨", "콜라", "맥주"])
print('시리즈 출력 :')
print('-'*15)
print(series)
```

```python
print('시리즈의 값 : {}'.format(series.values))
print('시리즈의 인덱스 : {}'.format(series.index))
```

```python
values = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
index = ['one', 'two', 'three']
columns = ['A', 'B', 'C']

df = pd.DataFrame(values, index=index, columns=columns)

print('데이터프레임 출력 ')
print('-'*25)
print(df)
```

```python
print('DataFrame의 인덱스 : {}'.format(df.index))
print('DataFrame의 열이름: {}'.format(df.columns))
print('DataFrame의 값 :')
print('-'*25)
print(df.values)
```

```python
data = [['2021/1', 18.25],['2021/2', 8.94],['2021/3', 8.40],['2021/4', 3.16],['2021/5', '1.97'],['2021/6', 3.30]
, ['2021/7', 5.43], ['2021/8', 4.17], ['2021/9', 3.77], ['2021/10', 2.37], ['2021/11', 3.47], ['2021/12', 1.95],]

df= pd.DataFrame(data)
print(df)
```
