```python
# 문자열 포매팅
# 숫자
"I eat %d apples." % 3
```

```python
# 문자열 포매팅
# 문자 (굳이 문자를 넣지 않아도 됨 실수 정수 가능)
"I eat %s apples." % "five"
```

```python
# 문자열 포매팅
# 변수
number = 3
"I eat %d apples" % number
```

```python
# 문자열 포매팅
# 2개 이상의 값

number = 10
day = "three"
"I ate %d apples. so I was sick for %s days." % (number, day)
```

```python
# 정렬과 공백
"%10s" % "hi" # %10s는 전체 길이가 10개인 문자열 공간에서 대입되는 값을 오른쪽으로 정렬하고 그 앞의 나머지는 공백으로 남겨 두라는 의미이다.
```

```python
"%-10sjane." % 'hi' # 왼쪽 정렬
```

```python
"%0.4f" % 3.42134234 # 소수점 4자리 까지 '.'은 소수점포인트 , 그 뒤의 숫자는 뒤에 나올 숫자 개수
```

```python
# format 함수 포매팅.  문자열의 format 함수를 사용하면 좀 더 발전된 스타일로 문자열 포맷 지정 가능
"I eat {0} apples".format(3) # 숫자
```

```python
"I eat {0} apples".format("five") # 문자열
```

```python
number = 3
"I eat {0} apples".format(number) # 변수
```

```python
number = 10
day = "three"
"I ate {0} apples. so I was sick for {1} days.".format(number, day) # 두 변수
```

```python
"I ate {number} apples. so I was sick for {day} days.".format(number=10, day=5) # 인덱스와 이름 혼용
```

```python
"{0:<10}".format("hi") # 왼쪽 정렬
```

```python
"{0:>10}".format("hi") # 오른쪽 정렬
```

```python
"{:^10}".format("hi") # 가운데 정렬
```

```python
"{0:=^10}".format('hi') # 공백 채우기1
```

```python
"{0:!<10}".format("hi") # 공백 채우기2
```

```python
y = 3.42134234
"{0:0.4f}".format(y) # 소수점 표현하기
```

```python
"{0:10.4f}".format(y) # 자리수 맞추기
```

```python
"{{ and }}".format() # { 또는 } 문자 표현
```

```python
# f 문자열 포매팅
name = "홍길동"
age = 30
f'나의 이름은 {name}입니다. 나이는 {age}입니다.'
```

```python
age = 30
f'나는 내년이면 {age+1}살이 된다.'
```

```python
d = {'name':'홍길동', 'age':30}
f'나의 이름은 {d["name"]}입니다. 나이는 {d["age"]}입니다.'
```

```python
f'{"hi":<10}' # 왼쪽 정렬
```

```python
f'{"hi":>10}' # 오른쪽 정렬
```

```python
f'{"hi":^10}' # 가운데 정렬
```

```python
f'{"hi":=^10}' # 공백채우기 1
```

```python
f'{"hi":!<10}' # 왼쪽 정렬하고 '!' 문자로 공백 채우기
```

```python
y = 3.42134234
f'{y:0.4f}'    # 소수점 4자리까지 표현
```

```python
f'{y:10.4f}'  # 소수점 4자리까지 표현하고 총 자리 수를 10으로 맞춤
```

```python
f'{{ and }}' # {} 쓰려면 동시 사용
```
