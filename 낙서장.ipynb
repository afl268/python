```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
matplotlib.rc('axes',unicode_minus = False)
matplotlib.rc('font',family = 'Malgun Gothic')
```

```python
df = pd.read_csv('./인천광역시_코로나19 확진자 현황_20220117.csv',encoding='cp949')
print(df)
```

```python
df['확진일'] = pd.to_datetime(df['확진일'],format='%Y-%m-%d')
```

```python
df12 = df.query('"2020"<= 확진일 <= "2020-12-31"')
print(df12)
```

```python
gyeyang12 = df12[df12['관리기관'] == '계양구']
gyeyang12.tail()
```

```python
plt.scatter( gyeyang12['인천'], gyeyang12['나이'], label = "확진자" , color = "red")

plt.legend(loc = "best")
plt.title('2020년도 계양구 확진자 나이분포')
plt.xlabel('확진자 수')
plt.ylabel('나이')
plt.show()
```
