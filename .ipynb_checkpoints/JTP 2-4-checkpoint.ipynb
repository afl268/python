```python
# 튜플과 리스트의 차이
# 리스트는 [] 으로 둘러싸지만 튜플은 ()으로 둘러싼다.
# 리스트는 그 값의 생성, 삭제, 수정이 가능하지만 튜플은 그 값을 바꿀 수 없다.

t1 = ()
t2 = (1,)
t3 = (1, 2, 3)
t4 = 1, 2, 3
t5 = ('a', 'b', ('ab', 'cd'))

# 값이 바뀌거나 변하지 않길 바란다면 튜플 사용
```

```python
t1 = (1, 2, 'a', 'b')
del t1[0]
# 튜플은 요소를 지우는 행위는 지원되지 않는다.
```

```python
t1 = (1, 2, 'a', 'b')
t1[0] = 'c'
# 요소값 변경 하려 하면 에러 발생
```

```python
# 튜플 인덱싱
t1 = (1, 2, 'a', 'b')
t1[0]
```

```python
t1[3] # t1[0], t1[3]처럼 인덱싱 가능
```

```python
# 튜플 슬라이싱
t1 = (1, 2, 'a', 'b')
t1[1:] # t1[1]부터 튜플의 마지막 요소까지 슬라이싱
```

```python
# 튜플 더하기
t1 = (1, 2, 'a', 'b')
t2 = (3, 4)
t1 + t2
```

```python
# 튜플 곱하기
t2 = (3, 4)
t2 * 3
```

```python
# 튜플 길이 구하기
t1 = (1, 2, 'a', 'b')
len(t1)
```
