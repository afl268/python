```python
def is_odd(number):
    if number % 2 == 1:
        return True
    else:
        return False
    
```

```python
is_odd(3)
```

```python
is_odd(4)
```

```python
def avg_number(*args):
    result = 0
    for i in args:
        result += i
    return result / len(args)

avg_number(1,2)
```
