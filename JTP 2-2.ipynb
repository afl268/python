```python
# 문자열에 작은따옴표 (') 포함시키기

food = "Python's favorite food is perl"
food
```

```python
a = '"Python is very easy." he says.' # 작은따옴표 안에 사용한 큰따옴표는 문자열을 만드는 기호로 인식되지 않는다.
a
```

```python
# 백슬래시(\)를 사용해서 작은따옴표와 큰따옴표 문자열에 포함시키기
food = 'Python\'s favorite food is perl'
say = "\"Python is very easy.\" he says."
food
```

```python
say
```

```python
# 줄바꿈을 할 때 \n을 삽입하는 방법이 있지만 읽기 불편하고 줄이 길어지는 단점이 있다.
multiline = "Life is too short\nYou need python"
multiline
```

```python
# 연속된 작은따옴표 3개(''') 또는 큰따옴표 3개(""") 사용하기
multiline = '''
Life is too short
You need Python
'''
multiline
```
