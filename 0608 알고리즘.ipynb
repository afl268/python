```python
"""a의 절대값을 구하는 과정 
    1. a가 0보다 크거나 같은지 확인, 만약 그렇다면 a를 결과로 돌려줌
    2. 1의 경우가 아니라면(a가 0보다 작으면) -a를 결과로 돌려줌
    3. a를 제곱하여 변수 b에 저장함.
    4. b의 제곱근을 구해 결과로 돌려줌
"""

import math

def abs_sign(a):
    if a >= 0:
        return a
    else:
        return -a
    
def abs_square(a):
    b = a * a
    return math.sqrt(b) # 수학 모듈의 제곱근 함수

print(abs_sign(5))
print(abs_sign(-3))
print(abs_square(3)) 
print(abs_square(-5)) 
```

```python
"""
함수사용

1부터 n까지의 합 구하기
이 문제에서 입력은 'n까지'에 해당하는 n이며 출력은 n=10이면 1부터 10까지의 합은 55, n = 100이면 1부터 100까지의 합은 5050이다.
이 55와 5050이 각각의 입력에 대한 출력이다.

def all_sum(b):
    sum = 0

    for i in range(1, b + 1):
        sum = sum + i
    return sum

print(all_sum(3)) 
"""

# 함수 미사용
a = int(input())
sum = 0

for i in range(1, a + 1):
    sum += i

print(sum)
```

```python
"""
어떤 알고리즘이 문제를 풀기 위해 해야 하는 계산이 얼마나 복잡한지 나타낸 정도를 '계산 복잡도' 라고 한다.
계산 복잡도를 표현하는 벙법에는 여러 가지가 있는데, 그 중 대문자 O(빅오) 표기법을 가장 많이 사용한다.
첫 번째 알고리즘은 입력 크기 n에 대해 사칙 연산(덧셈)을 n번 해야 한다. 이 때 이 알고리즘의 계산 복잡도를 O(n)이라 표현하게 된다.
입력 크기 n에 따라 덧셈을 두 번씩 하는 알고리즘을 O(2n)이라 표현할 수 있다고 생각하는데 이 때는 그냥 O(n)으로 표현한다.
대문자 O 표기법은 알고리즘에서 필요한 계산 횟수를 정확한 숫자로 표현하는 것이 아닌 입력 크기와의 관계로 표현하기 때문이다.
예를 들어 n이 10에서 20이 되면 계산 횟수가 입력 크기 n과 '정비례' 하기 때문에 O(n)이라 표기한다.

두 번째 알고리즘은 입력 크기 n과 무관하게 사칙연산을 3번 해야한다. 이 때 알고리즘의 계산 복잡도는 O(1)로 표현한다.
O(3)이 아닌 이유는 입력 크기 n과 필요한 계산의 횟수가 무관하다면, 모두 O(1)로 표현한다.

대문자 O 표기법은 알고리즘의 대략적인 성능을 표시하는 방법이다. 따라서 굉장히 세밀한 계산 횟수나 소요 시간을 표현한다기보다
입력 크기 n과 필요한 계산 횟수와의 '관계'에 더 주목하는 표현이라고 기억하면 된다.
"""
```

```python
""" 1부터 n까지 연속한 숫자의 제곱의 합을 구하는 프로그램을 for 반복문으로 만들어라."""

n = int(input())
sum = 0
for i in range(1, n + 1):
    sum = sum + i**2
print(sum)
```

```python
"""
첫 번째 숫자 17을 최대값으로 기억한다.
두 번째 숫자를 현재 최대값 17과 비교, 92가 17보다 크므로 최대값을 92로 바꿔 기억
세 번째 숫자 18을 현재 최대값 92와 비교, 18은 92보다 작으므로 지나간다
이후 반복
"""
"""
a = [17,92,18,33,58,7,323,42] # 리스트 선언
n = len(a) # 리스트 길이값 n 변수에 저장
max_v = a[0] # max_v 변수에 a 리스트의 첫 번째 값 저장
for i in range(1,n): # for문 선언
    
    if a[i] > max_v: # 만약 for문이 도는 동안 a[i] 값이 max_v(a[0])보다 크다면
        max_v = a[i] # 그 값을 저장
        
print(max_v)
"""

v = [17,92,18,33,58,7,323,42] # 리스트 선언
def find_max(a):
    n = len(a) # 리스트 길이값 n 변수에 저장
    max_v = a[0] # max_v 변수에 a 리스트의 첫 번째 값 저장
    
    for i in range(1, n): # for문 선언
        if a[i] > max_v: # 만약 for문이 도는 동안 a[i] 값이 max_v(a[0])보다 크다면
            max_v = a[i]  # 그 값을 저장
    return max_v 

print(find_max(v))
```

```python
"""
리스트에 숫자 n개 있을 때 가장 큰 값이 있는 위치 번호를 돌려주는 알고리즘을 만들어라.
"""
a = [17,92,18,33,58,7,323,42] # 리스트 선언
n = len(a) # 리스트 길이값 n 변수에 저장
max_v = 0 # 리스트 중 0번 위치를 최대값으로 기억
for i in range(1,n): # for문 선언
    
    if a[i] > a[max_v]: # 만약 이번 값이 현재까지 기억된 최대값보다 크면
        max_v = i # 최대값의 위치를 변경
         
print(max_v)
```

```python
"""
숫자 n개를 리스트로 입력받아 최소값을 구하는 프로그램을 만들어라.
"""

def min_v(a):
    n = len(a)
    min_v = a[0]

    for i in range(1,n):
        if a[i] < min_v:
            min_v = a[i]
    return min_v

v = [17,92,18,33,58,7,323,42]
print(min_v(v))
```
