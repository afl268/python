```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
matplotlib.rc('axes',unicode_minus = False)
matplotlib.rc('font',family = 'Malgun Gothic')
```

```python
df = pd.read_csv('C:/Users/emoot/data/incheon17213.csv',encoding='cp949')
df.head()
```

```python
subset = df[['일시','평균기온(℃)']]
subset1 = subset[0:12]
subset1

# csv에서 특정 열 추출 후 총 5년간의 데이터 중 1년치 행 추출
```

```python
subset1.plot()
plt.title("2017년 평균기온 표")
plt.xlabel("일시(월)")
plt.ylabel("평균기온")
plt.show()
```

```python
subset1.plot(color='#c02ad1') # 색상 지정
plt.title("2017년 평균기온 표")
plt.xlabel("일시(월)")
plt.ylabel("평균기온")
plt.show()
```

```python
subset1.plot(color='#c02ad1', linestyle= '--',linewidth=2) # 점선 그래프
plt.title("2017년 평균기온 표")
plt.xlabel("일시(월)")
plt.ylabel("평균기온")
plt.show()
```

```python
subset1.plot(color='#c02ad1', linestyle= 'solid',linewidth=2,
            marker='o',markersize=10,
            markeredgecolor='white', markeredgewidth=3,
            ) # 볼드선, 마커 그래프
plt.title("2017년 평균기온 표")
plt.xlabel("일시(월)")
plt.ylabel("평균기온")
plt.show()
```
