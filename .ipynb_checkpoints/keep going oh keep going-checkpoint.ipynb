```python
# https://linuxtut.com/en/cea310b2c36a01b970a6/

import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

x = np.linspace(0,2*np.pi,100)
for i in range(30):   
    plt.plot(x,i*np.sin(x),color=cm.jet(i/30.0))
plt.xlim(0,2*np.pi)
plt.savefig("cm.jpg")
plt.show()
```
