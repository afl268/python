```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
import numpy as np
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
matplotlib.rc('axes',unicode_minus = False)
matplotlib.rc('font',family = 'Malgun Gothic')
```

```python
df = pd.read_csv('C:/Users/emoot/data/is1week.csv',encoding='cp949')
print(df)
```

```python
df["일시"] = pd.to_datetime(df["일시"])
df.head(5)
```

```python
spot_ = "인천"
d_ = df[(df["지점명"]==spot_) & (df["일시"].dt.month==1)]
plt.figure(figsize=(10,5))
plt.title("1월 %s 지점 기온 추이" % spot_, fontsize=15)
plt.plot(d_["일시"], d_["기온(°C)"], "-", color='black', label=str(spot_))
plt.grid()
plt.legend(fontsize=13)
plt.xticks(rotation=90)
plt.show()
```

```python
spot_ = "인천"
d_ = df[(df["지점명"]==spot_) & (df["일시"]<"2022-01-20")]
plt.figure(figsize=(10,5))
plt.title("시간대별 기온 추이(2022-01-20)", fontsize=15)
ax = plt.subplot(1,1,1)
ax.plot(d_["일시"],d_["기온(°C)"], "-o", color='red', label=str(spot_))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
plt.legend(fontsize=13)
plt.grid()
plt.show()
```

```python
spot_ = "인천"
d_ = df[(df["지점명"]==spot_) & (df["일시"].dt.month==1)]
plt.figure(figsize=(12,6))
plt.title("1월 %s 지점 기온,풍속 추이" % spot_, fontsize=15)
plt.xticks(rotation=90)
ax1 = plt.subplot(111)
ax2 = ax1.twinx()
a, =  ax1.plot(d_["일시"], d_["기온(°C)"], "-", color='red', label="기온(°C)")
b, = ax2.plot(d_["일시"], d_["풍속(m/s)"], "-", color='blue', label="풍속(m/s)")
p = [a, b]
ax1.legend(p, [p_.get_label() for p_ in p])
plt.show()
```
