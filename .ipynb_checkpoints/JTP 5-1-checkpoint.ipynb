```python
# 객체에 숫자 지정
class FourCal:
    def setdata(self, first, second): # 메서드의 매개변수
        self.first = first # 메서드의 수행문
        self.second = second # 메서드의 수행문
# pass 문장 삭제, setdata 함수 생성 , 클래스 안에 구현된 함수는 메서드라 부름
```

```python
# 곱, 빼, 나누기 기능
class FourCal:
    def setdata(self, first, second): 
        self.first = first 
        self.second = second 
    def add(self):
        result = self.first + self.second
        return result
    def mul(self):
        result = self.first * self.second
        return result
    def sub(self):
        result = self.first - self.second
        return result
    def div(self):
        result = self.first / self.second
        return result
```

```python
a = FourCal()
a.setdata(4, 2)
a.add()
```

```python
a.mul()
```

```python
a.sub()
```

```python
a.div()
```

```python
# 생성자 : 객체가 생성될 때 자동으로 호출되는 메서드
# 객체에 초기값을 설정해야 할 때 setdata 메서드 호출하는 것보다 생성자를 구현하는게 더 안전
class FourCal:
    def __init__(self, first, second): 
        self.first = first 
        self.second = second 
    def add(self):
        result = self.first + self.second
        return result
    def mul(self):
        result = self.first * self.second
        return result
    def sub(self):
        result = self.first - self.second
        return result
    def div(self):
        result = self.first / self.second
        return result
```

```python
a = FourCal() # 생성자의 매개변수에 해당하는 값이 전달되지 않음
```

```python
a = FourCal(4, 2)
a.add()
```

```python
# 클래스 상속
class  MoreFourCal(FourCal):
    pass
a = MoreFourCal(4, 2)
a.add()
```

```python
# 상속 이유 : 기존 클래스 변경 없이 기능 추가, 기능 변경하고자 할 때 사용
class  MoreFourCal(FourCal):
    def pow(self):
        result = self.first ** self.second
        return result

a = MoreFourCal(4, 2)
a.pow()
```

```python
# 메서드 오버라이딩 : 부모클래스에 있는 메서드를 동일한 이름으로 다시 만드는 것
class SafeFourCal(FourCal):
    def div(self):
        if self.second == 0:
            return 0
        else :
            return self.first / self.second
```

```python
a = SafeFourCal(4, 0)
a.div()
```
