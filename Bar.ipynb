```python
import matplotlib
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
matplotlib.rc('axes',unicode_minus = False)
matplotlib.rc('font',family = 'Malgun Gothic')
```

```python
## 데이터
age_category = ['10대 이하', '20대', '30대', '40대', '50대', '60대 이상'] ## 연령 카테고리, x축 눈금에 표시될 라벨
num_patient = [3,3,5,7,8,9] ## 방문환자수
```

```python
## 시각화
plt.figure(figsize=(10,10)) ## Figure 생성 사이즈는 10 by 10
xtick_label_position = list(range(len(age_category))) ## x축 눈금 라벨이 표시될 x좌표
plt.xticks(xtick_label_position, age_category) ## x축 눈금 라벨 출력
 
plt.bar(xtick_label_position, num_patient) ## 바차트 출력 (수직)
 
plt.title('지난달 연령별 방문환자 수',fontsize=20) ## 타이틀 출력
plt.xlabel('연령') ## x축 라벨 출력
plt.ylabel('방문환자수') ## y축 라벨 출력
plt.show()
```

```python
plt.figure(figsize=(10,10)) ## Figure 생성 사이즈는 10 by 10
ytick_label_position = list(range(len(age_category))) ## y축 눈금 라벨이 표시될 x좌표
plt.yticks(ytick_label_position, age_category) ## y축 눈금 라벨 출력
 
plt.barh(ytick_label_position, num_patient) ## 바차트 출력 (수평)
 
plt.title('지난달 연령별 방문환자 수',fontsize=20) ## 타이틀 출력
plt.ylabel('연령') ## y축 라벨 출력
plt.xlabel('방문환자수') ## x축 라벨 출력
plt.show()
```

```python
plt.figure(figsize=(10,10)) ## Figure 생성 사이즈는 10 by 10
colors = sns.color_palette('cool',len(age_category)) ## 색상 지정
ytick_label_position = list(range(len(age_category))) ## y축 눈금 라벨이 표시될 x좌표
plt.yticks(ytick_label_position, age_category) ## y축 눈금 라벨 출력
 
plt.barh(ytick_label_position, num_patient, color = colors, edgecolor=colors ,alpha=0.8, linewidth=5) ## 바차트 출력
 
plt.title('지난달 연령별 방문환자 수',fontsize=20) ## 타이틀 출력
plt.ylabel('연령') ## y축 라벨 출력
plt.xlabel('방문환자수') ## x축 라벨 출력
plt.show()

# edgecolor(테두리) 색 종류 : https://matplotlib.org/3.2.1/gallery/color/named_colors.html
```

```python
# How to add text on a bar with matplotlib ?: https://moonbooks.org/Articles/How-to-add-text-on-a-bar-with-matplotlib-/

tick_size = 10 ## 눈금 폰트 사이즈
axis_label_size = 15 ## x축, y축 폰트 사이즈

fig = plt.figure(figsize=(10,10)) ## Figure 생성 사이즈는 10 by 10
ax = fig.add_subplot() ## Axes 추가
colors = sns.color_palette('cool',len(age_category)) ## 색상 지정
ytick_label_position = list(range(len(age_category))) ## y축 눈금 라벨이 표시될 x좌표
plt.yticks(ytick_label_position, age_category) ## y축 눈금 라벨 출력
 
barhs = plt.barh(ytick_label_position, num_patient, color = colors, edgecolor=colors ,alpha=0.8, linewidth=5) ## 바차트 출력

for i, barhs in enumerate(barhs):
    ax.text(0.85 * barhs.get_width(), barhs.get_y() + barhs.get_height() / 2.4
            , num_patient[i], ha='right', fontsize=13)
    
plt.title('지난달 연령별 방문환자 수',fontsize=20) ## 타이틀 출력
plt.ylabel('연령') ## y축 라벨 출력
plt.xlabel('방문환자수') ## x축 라벨 출력
plt.show()
```
