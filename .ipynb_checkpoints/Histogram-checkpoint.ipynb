```python
plt.hist(x, 20, histtype = 'barstacked', color = 'r', edgecolor = 'b')
# histtype에 들어갈 수 있는 값의 종류 : bar , barstacked, step, stepfilled
 
plt.title('Histogram')
plt.show()
```

```python
np.random.seed(1)
 
## 두개의 데이터 생성
mean, sigma = 5, 1
x = mu + sigma * np.random.randn(100)
mean_y, sigma_y = 10, 1.5
y = mu_y + sigma_y * np.random.randn(100)
 
fig = plt.figure(figsize=(8,8))
fig.set_facecolor('white')
 
plt.hist(x, 10, density=True, color='g', alpha=0.75, edgecolor='k') # y축에 비율 나타내기 위해 density = True로 지정
plt.hist(y, 10, density=True, color='b', alpha=0.75, edgecolor='k')
 
plt.title('Histogram')
plt.show()
```

```python
np.random.seed(1)
 
## 두개의 데이터 생성
mean, sigma = 5, 1
x = mu + sigma * np.random.randn(100)
mean_y, sigma_y = 10, 1.5
y = mu_y + sigma_y * np.random.randn(100)
 
fig = plt.figure(figsize=(8,8))
fig.set_facecolor('white')

# 스택 히스토그램을 그릴 데이터를 하나의 리스트에 담아서 hist의 첫 번째 인자로 넘김
# 이 때 색상을 지정할 때는 데이터 리스트와 똑같은 길이의 색상 리스트를 지정해주어야 함
plt.hist([x,y], 20, 
         density=True, 
         color=['g','b'], 
         histtype='barstacked', 
         alpha=0.75, 
         edgecolor='k')
 
plt.title('Histogram')
plt.show()
```
