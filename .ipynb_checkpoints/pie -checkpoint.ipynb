```python
# 출처 : https://zephyrus1111.tistory.com/

## 데이터 준비
labels = ['Apple','Banana','Grape','Pear','Peach'] ## 라벨
frequency = [20,20,380,240,200] ## 빈도
```

```python
import matplotlib.pyplot as plt
 
fig = plt.figure(figsize=(8,8)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 색상 하얀색
 
ax = fig.add_subplot() ## 프레임 생성
 
ax.pie(x=frequency,
       labels=labels,
       autopct=lambda p : '{:.2f}%'.format(p)) ## 파이 차트 출력
plt.show() # pie 차트의 처음 시작은 3시 방향부터 시작해서 반시계 방향으로 돌음
```

```python
fig = plt.figure(figsize=(8,8)) 
fig.set_facecolor('white') 
 
ax = fig.add_subplot() 
 
ax.pie(x=frequency,
       labels=labels,
       startangle=90, ## 시작점을 90도로 지정
       autopct=lambda p : '{:.2f}%'.format(p)) 
plt.show() 
```

```python
fig = plt.figure(figsize=(8,8)) 
fig.set_facecolor('white') 
 
ax = fig.add_subplot() 
 
ax.pie(x=frequency,
       labels=labels,
       startangle=90, ## 시작점을 90도로 지정
        counterclock=True, ## 반시계 방향으로 파이차트를 그림
       autopct=lambda p : '{:.2f}%'.format(p)) ## 파이 차트 출력
plt.show() 
```

```python
fig = plt.figure(figsize=(8,8)) 
fig.set_facecolor('white') 
 
ax = fig.add_subplot() 
 
ax.pie(x=frequency,
       labels=labels,
       startangle=90, ## 시작점을 90도로 지정
        counterclock=False, ## 시계 방향으로 파이차트를 그림
       autopct=lambda p : '{:.2f}%'.format(p)) ## 파이 차트 출력
plt.show() 
```

```python
## 데이터 준비
labels = ['Apple','Banana','Grape','Pear','Peach'] ## 라벨
frequency = [120,120,380,240,200] ## 빈도

fig = plt.figure(figsize=(8,8)) 
fig.set_facecolor('white') 
 
ax = fig.add_subplot() 

explode = [0.1,0.1,0,0,0] ## 튀어나오는 정도를 결정한다.
ax.pie(x=frequency,
       labels=labels,
       explode = explode,## Apple과 Banana를 튀어나오게한다.
       startangle=90, 
        counterclock=False, 
       autopct=lambda p : '{:.2f}%'.format(p)) 
plt.show() 
```

```python
fig = plt.figure(figsize=(8,8)) 
fig.set_facecolor('white') 
 
ax = fig.add_subplot() 


ax.pie(x=frequency,
       labels=labels,
       startangle=90,
        counterclock=False, 
       autopct=lambda p : '{:.2f}%'.format(p)) 
plt.legend() ## 범례 표시
plt.show() 
```

```python
fig = plt.figure(figsize=(8,8))
fig.set_facecolor('white') 
ax = fig.add_subplot()
 
pie = ax.pie(frequency, # legend함수에 파이정보 넘기기 위해 pie 변수에 ax.ple가 리턴하는 값 삽입
             # 원 바깥쪽의 라벨을 표시하지 않기 위해 labels인자 삭제
       startangle=90, 
       counterclock=False, 
       autopct=lambda p : '{:.2f}%'.format(p)
       )
 
plt.legend(pie[0],labels) # 그래프 요소가 있는 frequency가 1번째에 있기에 pie[0]를 legend 함수 첫 인자에 넣음
                          # 범례에 라벨을 표시하기 위해 두 번째 인자에는 labels를 넣음
plt.show()
```

```python
fig = plt.figure(figsize=(8,8))
fig.set_facecolor('white') 
ax = fig.add_subplot()
 
pie = ax.pie(frequency,
       startangle=90, 
       counterclock=False, 
       autopct=lambda p : '{:.2f}%'.format(p) ,
       colors = ['red','yellow','purple','aqua','skyblue'] ## 색상 지정
       )
 
plt.legend(pie[0],labels) 
                          
plt.show()
```

```python
import seaborn as sns # seaborn을 이용해 색상 넣기

colors = sns.color_palette('hls',len(labels)) ## 라벨의 개수 만큼 색상 리스트 생성 
 
fig = plt.figure(figsize=(8,8)) 
fig.set_facecolor('white') 
ax = fig.add_subplot() 
 
plt.legend(pie[0],labels)  

pie = ax.pie(frequency, 
       startangle=90, 
       counterclock=False, 
       autopct=lambda p : '{:.2f}%'.format(p),
       colors = colors ## 색상 지정
       )
```
