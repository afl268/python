```python
# 만약 3000원 이상의 돈을 가지고 있으면 택시를 타고 그렇지 않으면 걸어가라
money = 2000
if money >= 3000:
    print("택시")
else:
    print("걷기")
```

```python
money = 2000
card = True
if money >= 3000 or card:
    print("택시")
else:
    print("걷기")
```

```python
pocket = ['paper', 'cellphone', 'money']
if 'money' in pocket:
    print("택시")
else:
    print("걷기")
```

```python
# 조건문에서 아무 일도 하지 않게 설정
pocket = ['paper', 'cellphone', 'money']
if 'money' in pocket:
    pass
else:
    print("걷기")
```

```python
# elif
pocket = ['paper', 'handphone']
card = True
if 'money' in pocket:
    print("택시")
elif card:
    print("택시")
else:
    print("걷기")
```
