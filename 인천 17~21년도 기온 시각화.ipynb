```python
df = pd.read_csv('C:/Users/emoot/data/incheon17213.csv',encoding='cp949')
print(df)

# csv 불러오기, 인코딩은 cp949
```

```python
subset = df[['일시','평균기온(℃)']]
subset
# csv에서 특정 열 추출
# 행 열 반전

# 일시 : 기록일시
```
