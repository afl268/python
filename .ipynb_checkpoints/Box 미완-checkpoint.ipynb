```python
# 출처 : https://wikidocs.net/141959

import matplotlib.pyplot as plt
import numpy as np

# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
# 3. 그래프 그리기
fig, ax = plt.subplots()

ax.boxplot([data_a])
ax.set_ylim(-10.0, 10.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

plt.show()
```

```python
# 출처 : https://blog.naver.com/qbxlvnf11/221570541564
import pandas as pd
import seaborn as sns

tips = sns.load_dataset('tips')
print(tips.shape)
print(tips.head)
```

```python
box = dict(markerfacecolor='g', marker='D') # marker color : green , marker shape : diamond
plt.boxplot(tips['tip'], flierprops=box)
plt.title("tips box plot")
plt.show()
```

```python
green_diamond = dict(markerfacecolor='g', marker='D')
plt.boxplot(tips['tip'], vert=False, flierprops=green_diamond) # vert(turn) -> False
plt.title("tips box plot")
plt.show()
```

```python
data = [tips['tip'], tips['total_bill']]

box = dict(markerfacecolor='r', marker='s')
plt.boxplot(data, flierprops=box)
plt.title("multiple box plot")
plt.show()
```

```python
import plotly.express as px
df = px.data.tips()
fig = px.box(df, y="total_bill")
fig.show()
```
