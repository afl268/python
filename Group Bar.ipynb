```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Type A

barWidth = 0.25
one = [30,100,30,25]
two = [50,20,40,100]

plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 12
plt.rcParams["figure.figsize"] = (12, 8)

r1 = np.arange(len(one)) # 리스트 one
r2 = [x + barWidth for x in r1]

plt.bar(r1, one, color='#0080ff', width=barWidth, edgecolor='#0075b1',linewidth=2,  label='이진' )
plt.bar(r2, two, color='#557f2d', width=barWidth, edgecolor='#007a00', linewidth=2, label='미지' )
 
plt.xlabel('과목', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(one))], ['국어', '수학', '영어', '과학'])
plt.title('성적')
 
plt.legend()
plt.show()
```

```python
# Type B

R1 = [70, 54, 12]
R2 = [10, 108, 87] 
R3 = [26, 38, 66]


label = ['1', '2', '3']

plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 12
plt.rcParams["figure.figsize"] = (12, 8)

plt.figure()

x = np.arange(len(label))

plt.bar(x-0.0, R1, label='랜덤1', width=0.2, color='#ed7d95', edgecolor='#0075b1', linewidth=2,)
plt.bar(x+0.2, R2, label='랜덤2', width=0.2, color='#FF5800', edgecolor='#007a00', linewidth=2)
plt.bar(x+0.4, R3, label='랜덤3', width=0.2, color='#d81c2f', edgecolor='#000000', linewidth=2)
plt.xticks(x, label)

plt.legend()
plt.xlabel('??')
plt.ylabel('숫자')
plt.ylim(0, 110)
plt.title('3개')

plt.show()
plt.close()
```
