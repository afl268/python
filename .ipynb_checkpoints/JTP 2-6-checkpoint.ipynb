```python
# 집합(set)은 파이썬 2.3부터 지원하기 시작한 자료형으로, 집합에 관련된 것을 쉽게 처리하기 위해 만든 자료형

s1 = set([1,2,3,])
s1
```

```python
s2 = set("Hello")
s2 # 중복허용 없음, 순서가 없음
```

```python
s1 = set([1,2,3])
l1 = list(s1)
l1
```

```python
l1[0]
```

```python
t1 = tuple(s1)
t1
```

```python
t1[0]
```

```python
s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])
```

```python
# 교집합
s1 & s2
```

```python
s1.intersection(s2)
```

```python
# 합집합
s1 | s2
```

```python
s1.union(s2)
```
